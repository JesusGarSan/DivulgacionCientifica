```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

```

```python
# Ley de Snell
# n_1 * sin(O_1) = n_2 * sin(O_2)
def snell(n_1, n_2, O_1):
    if abs(n_1/n_2 * np.sin(O_1/180*np.pi) )>1: return False
    return np.arcsin( n_1/n_2 * np.sin(O_1/180*np.pi) ) * 180/np.pi
```

```python
O_1 = 60 # º
n_1 = 4
n_2 = 1.4
O_2 =  snell(n_1, n_2, O_1) 

if O_2: print(O_2)
else: print('Refracción Interna Total')

longitud_rayos = 1

# rayo incidente
inc_x = [-longitud_rayos * np.sin(np.pi*O_1/180),0]
inc_y = [longitud_rayos * np.cos(np.pi*O_1/180),0]


# rayo refractado
refrac_x = [longitud_rayos * np.sin(np.pi*O_2/180),0]
refrac_y = [-longitud_rayos * np.cos(np.pi*O_2/180),0]


```

```python
fig, ax = plt.subplots(figsize=(12,10))
ax.axis('off')
ax.fill_between([-1,1], [1,1], color='white') # medio 1
ax.fill_between([-1,1], [-1,-1], color='cyan') # medio 2
ax.vlines(0, -1, 1, color='grey', linestyles='dashed') # vertical
plt.plot(inc_x, inc_y, color='red')
if O_2: plt.plot(refrac_x, refrac_y, color='orange')
```
